````
#QUESTION 5) -- HOW MANY TIMES "CHEAP" WORD USED IN SEARCH OUT OF HOW MUCH ..
````

````
cheap_used = 0 
l1 = str('Cheap')
for i in flight_all['title'] : 
  if l1 in i :
    cheap_used = cheap_used + 1 
  else :
    pass 
print(cheap_used)
````

````
total = flight_all['title'].count()
print(total)
````

````
not_used_cheap = total - cheap_used 
print(not_used_cheap)
````

````
labels = ['total','cheap_used','not_used_cheap']
values = [total , cheap_used , not_used_cheap]
plt.figure(figsize=(6,4), dpi=100)
bars = plt.bar(labels, values)

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))


plt.show()
````
